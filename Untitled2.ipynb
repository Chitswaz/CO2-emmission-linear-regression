```python
df = pd.read_csv("china_gdp.csv")
```

```python
# open the csv file
df.head()
```

```python
plt.plot(df.Year, df.Value)
plt.show()
```

```python
# approximating with the logistic function
x = np.arange(-5.0, 5.0, 0.1)
y = 1.0 / (1 + np.exp(-x))
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
```

```python
def sigmoid(x, B1, B2):
  return 1/ ( 1 + np.exp( -B1*(x - B2)) )
```

```python
B1 = 0.10
B2 = 1990.0
y_pred = sigmoid(df.Year, B1, B2)
plt.plot(df.Year, y_pred)
plt.show()
```

```python
# normalizing the data
x_data = df.Year / max(df.Year)
y_data = df.Value / max(df.Value)
```

```python
from scipy.optimize import curve_fit
optimized, covariance = curve_fit(sigmoid, x_data, y_data)
print(" beta_1 = %f, beta_2 = %f" % (optimized[0], optimized[1]))
```

```python
from sklearn.metrics import r2_score
x_new = np.linspace(1960, 2020, 55)
x_new_norm = x_new / max(x_new)
y_new =12000000000000 * sigmoid(x_new_norm, *optimized)
plt.figure(figsize=(8,5))
plt.plot(df.Year, df.Value, 'b', label='data')
plt.plot(x_new, y_new, 'r', label='fit')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend()
plt.show()
print( 'accuracy:', r2_score(df.Value, y_new)*100 )
```
